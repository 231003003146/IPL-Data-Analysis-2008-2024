#### IPL (2008-2024) Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
scorecard = pd.read_csv('deliveries.csv')
```

```python
scorecard.head(5)
```

#### ANALYSIS 1:Top Run Scorers in ipl History

```python
top_batsman = scorecard.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
top_batsman
```

```python
plt.figure(figsize=(11,5))
top_batsman.plot(kind='bar')
plt.title("Top Batsman in Ipl History")
plt.ylabel("RUNS")
plt.show()
```

### ANALYSIS 2:Top Wicket Takers in IPL History

```python
top_bowler = scorecard[scorecard['is_wicket'] == 1]['bowler'].value_counts().head(10)
```

```python
top_bowler
```

```python
plt.figure(figsize=(11,5))
top_bowler.plot(kind='bar')
plt.title('Top Wicket taker bowler in IPL History')
plt.ylabel('WICKET')
plt.show()
```

```python
top_bowler = top_bowler.reset_index()
```

```python
top_bowler
```

```python
plt.figure(figsize=(11,5))
sns.barplot(data=top_bowler,
            x = 'bowler',
            y = 'count',
            hue='bowler',
            palette='tab10',
            legend=False
            )
plt.title('Top Wicket taker bowler in IPL History')
plt.ylabel('WICKET')
plt.show()
```
